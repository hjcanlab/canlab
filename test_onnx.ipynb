```python
import onnx

# 1. ONNX 모델 불러오기
onnx_path = "/workspace/testcode/models copy/thermal_skip_480x640.onnx"
model = onnx.load(onnx_path)

# 2. 모델 구조 검증 (optional)
onnx.checker.check_model(model)

# 3. 그래프 요약 출력
print("==== Model Graph Info ====")
print("입력:")
for inp in model.graph.input:
    print(" -", inp.name, [dim.dim_value for dim in inp.type.tensor_type.shape.dim])

print("출력:")
for out in model.graph.output:
    print(" -", out.name, [dim.dim_value for dim in out.type.tensor_type.shape.dim])

print("\n==== Initializers (Weights) ====")
print(f"총 weight 텐서 개수: {len(model.graph.initializer)}")
for i, init in enumerate(model.graph.initializer[:10]):  # 앞부분 10개만 출력
    print(f"[{i}] 이름={init.name}, shape={[d for d in init.dims]}, dtype={init.data_type}")

print("\n(참고) 전체 weight 이름은 model.graph.initializer 리스트에 들어있습니다.")
```
